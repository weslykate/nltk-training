```
import numpy as np

a = np.array([[22, 5, 16, 9], 
              [13, 20, 19, 0],
              [5, 9, 1, 16],
              [27, 13, 15, 19]])
print('a\n', a)

b = np.array([[10, 5, 20, 20], 
              [5, 21, 22, 27],
              [26, 19, 22, 3],
              [27, 8, 18, 7]])

print('b\n', b)

c = np.array([24, 3, 17, 7, 11, 2, 21, 2, 24, 18, 3, 11, 14, 25, 2, 20])
d = np.array([3, 15, 24, 22, 19, 24, 4, 29, 6, 3, 20, 1, 11, 25, 26, 6])

print('c\n', c)
print('d\n', d)


print('\nAddition\n', a + b)
print('\nSubtraction\n', a - b)
print('\nMultiplication\n', a * b)
print('\nMatrix Multiplication\n', np.dot(a, b))

print('\nAdd\n', c+d)
print('\nSubtract\n', c-d)
print('\nMultiplication\n', c*d)
print('\nMatrix Multiplication\n', np.dot(c, d))

print('\nA + 10 \n', a + 10)
print('\nB + 10 \n', b + 10)
print('\nc + 15 \n', c + 15)
print('\nd + 15 \n', d + 15)

print('\n(2 * A) + (2 * B)\n', np.multiply(2, a) + np.multiply(2, b))
print('\n(A + B) * 3\n', (a + b) * 3)
print('\n(A - B + A - B + B) * 4 \n', (a - b + a - b + b) * 4)
```
